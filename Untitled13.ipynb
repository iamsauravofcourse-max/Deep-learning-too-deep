Tensor = speclized data storing structure designed to holder higher dimension data and for better computing efficiency

Types of Tensor
1. Scalar tensor = it is single values, like we calculate loss function using actual value - model prediction and we get loss that is a single value. Examples: 5, 6.07 etc

2. Vector(1D Tensor) = it is simply a row. we can take word embedding example of it, like HELLO can be represented in number form and from that we can get something like [23, 13, 56] etc

3. Matrices(2D Tensor) = we can take the example of MNIST dataset in which we have a text and it has one value from (0, 255) and it has width and height, so we get (120, 560, 24) etc

4. 3D tensor = example being rgb, there are three layers of red, green and blue and like the matrices we have 3 layers for this

5. 4D tensor = we can take example of images of this in sequence like we take 32 images in sequence

6. 5D tensor = video data there are different frame with the image which has values and there are batches of video

Pytorch Autograd

```python
import math
```

```python
def dy_dz(x):
  return 2 * x + math.cos(x**2)
dy_dz(2)
```

```python
def dy_dx(x,y):
  return (2.718**y) * (math.cos(y)/2*x) * (2*x)
dy_dx(2,3)
```

```python
x = torch.tensor(2.0, requires_grad = True)
y = x ** 2
z = torch.sin(y)
z.backward()
x.grad ## it simply doing backward chain rule so it is find dz/dx, it can be only find for the root not leaf
```

```python
## you can do similarly for vector as well
x = torch.tensor([1,2,3], requires_grad=True, dtype=torch.float)
```

```python
z = torch.mean(x)
```

```python
z.backward()
x.grad # we are finding all the partial derivate of each point like x1, x2 and x3 for each of them
```

```python
x.grad.zero_() #after each epooch we need to make the it zero otherwise it will acculumate
```

```python
x.requires_grad_(False) #if there is a time where you don't need backward and want to show the values of x as it is then you do this like after the end of training of neural
# you can also use torch_no_grad or detach for the same effect
```
